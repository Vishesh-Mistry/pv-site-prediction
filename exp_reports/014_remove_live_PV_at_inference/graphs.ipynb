```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
data = pd.DataFrame({
    'Model': ['Generic w/o Live PV', 'Bespoke w/o Live PV', 'Generic w Live PV', 'Bespoke w Live PV'],
    'Error': [12.00, 11.96, 9.98, 9.95]
})

# Define colors for each model
colors = ['#4C9A8E' if 'Generic' in model else '#FFD053' for model in data['Model']]

# Set a more visually pleasing font
plt.rcParams['font.family'] = 'sans-serif'


# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Error', data=data, palette=colors)
plt.title('Error Comparison for Different Models', fontsize=14)
plt.xlabel('Model')
plt.ylabel('Avereage Error % (MAE/Capacity)')
plt.ylim(9.5, 12.5)  # Set the y-axis limits for better visualization

# Display the values on top of the bars
for index, value in enumerate(data['Error']):
    plt.text(index, value + 0.1, str(value), ha='center', va='bottom', fontsize=17)

# Show the plot
plt.show()
```

```python
# Data
pv_dropout = [0.1, 0.2, 0.3]
error = [0, -0.24, 0.60]
error_live = [0, 0.14, 0.41]

# Create a plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the lines with different colors and labels
ax.plot(pv_dropout, error, marker='o',markersize=12, color='#086788', label='No Live PV')
ax.plot(pv_dropout, error_live, marker='o',markersize=12, color='#FFAC5F', label='Live PV')

# Set the title and labels
ax.set_title('Error vs PV Dropout', fontsize=16)
ax.set_xlabel('PV Dropout', fontsize=14)
ax.set_ylabel('% Change in Error (from PV Dropout = 0.1)', fontsize=14)

# Set a more visually pleasing font
plt.rcParams['font.family'] = 'sans-serif'

# Set x-axis ticks every 0.5
ax.set_xticks([0.1, 0.15, 0.2, 0.25, 0.3])


# Add a legend
ax.legend(fontsize=14)

# Show the plot
plt.show()

fig.savefig('pv_dropout_change.png')
```
